## Memory optimization 

```python
# Downcast numbers to smaller types
Flights_data = Flights_data.apply(pd.to_numeric, errors='ignore', downcast='integer')
Flights_data = Flights_data.apply(pd.to_numeric, errors='ignore', downcast='float')

# Convert object/text columns to category for shrinkage
for col in Flights_data.select_dtypes(include=['object']):
    if Flights_data[col].nunique() < len(Flights_data) * 0.5:   # convert only if beneficial
        Flights_data[col] = Flights_data[col].astype('category')

# Save as compressed CSV (strong compression)
Flights_data.to_csv("airlines_flights_data_compressed.csv.gz", index=False, compression='gzip')

# Load your file
df = pd.read_csv("C:\\Users\\Rishik Reddy\\OneDrive\\Desktop\\AirData\\DataSet\\airlines_flights_data.csv")

# Downcast numbers to smaller types
df = df.apply(pd.to_numeric, errors='ignore', downcast='integer')
df = df.apply(pd.to_numeric, errors='ignore', downcast='float')

# Convert object/text columns to category for shrinkage
for col in df.select_dtypes(include=['object']):
    if df[col].nunique() < len(df) * 0.5:   # convert only if beneficial
        df[col] = df[col].astype('category')

# Save as compressed CSV (strong compression)
df.to_csv("your_file_compressed.csv.gz", index=False, compression='gzip')
```

## Delhi Airport flights details according to Shift-wise 

```python
from IPython.display import display, Markdown

delhi_flights = Flights_data[Flights_data['source_city'] == 'Delhi']

# Separate the data by departure_time categories
early_morning_flights = delhi_flights[delhi_flights['departure_time'] == 'Early_Morning']
morning_flights = delhi_flights[delhi_flights['departure_time'] == 'Morning']
afternoon_flights = delhi_flights[delhi_flights['departure_time'] == 'Afternoon']
evening_flights = delhi_flights[delhi_flights['departure_time'] == 'Evening']
night_flights = delhi_flights[delhi_flights['departure_time'] == 'Night']

# Display each DataFrame for the user
display(Markdown("### Early Morning Flights from Delhi"))
display(early_morning_flights)

display(Markdown("### Morning Flights from Delhi"))
display(morning_flights)

display(Markdown("### Afternoon Flights from Delhi"))
display(afternoon_flights)

display(Markdown("### Evening Flights from Delhi"))
display(evening_flights)

display(Markdown("### Night Flights from Delhi"))
display(night_flights)
```

```python
import pandas as pd

Flights_data = pd.read_csv(r"C:\Users\Rishik Reddy\OneDrive\Desktop\AirData\DataSet\airlines_flights_data.csv")

delhi_flights = Flights_data[Flights_data['source_city'] == 'Delhi']

# Take only the departure_time column
delhi_departure_times = delhi_flights['departure_time']

# Get how many flights are in each time category
shift_counts = delhi_departure_times.value_counts()

# Display the result
print(shift_counts)
```

## Graphical representation of Total Flights 

```python
import matplotlib.pyplot as plt
```

```python
#bar chart for departing flights from each city

city_counts = Flights_data['source_city'].value_counts()

plt.figure(figsize=(10,6))
city_counts.plot(kind='bar')

plt.title('Number of Departing Flights From Each City')
plt.xlabel('City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for Airlines

Flights_data['airline'].value_counts().plot(kind='pie', figsize=(10,10), autopct='%1.1f%%',startangle=120)
plt.title('Number of Flights per Airline')
plt.show()
```
